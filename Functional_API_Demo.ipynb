```python
from tensorflow.keras.models import Model # Functional API
```

```python
model = Model(inputs =x , outputs =[output1,output2] )
```

```python
from tensorflow.keras.layers import *
```

```python
x = Input(shape = (3,)) # because 3 columns
```

```python
hidden1 = Dense(128,activation='relu')(x)
hidden2 = Dense(64,activation='relu')(hidden1)

output1 = Dense(1,activation='linear')(hidden2) # becuase predict age (linear output)
output2 = Dense(1,activation='sigmoid')(hidden2) # because predict city (binary output)
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(model,show_shapes=True)
```
